```python
combinations = []
populations = [25,50,75,100,150,200,500,1000,1500,2000,2500,5000,10000]
for prefix, aa_seq, tissue in zip(prefixes, proteins, tissues):
    for population in populations:
        combinations.append((prefix,aa_seq,tissue,population))
```

```python
def multi_GA_wrapper(args):
    return run_GA_comb(*args)

def run_GA_comb(prefix,aa_seq,tissue,population):
    print(prefix+str(population),'\n')
    filename = prefix+str(population)+'.csv'
    timefit = run_pygad_variable_pop_mp.run_GA(aa_seq, 
           tissue, 
           filename, 
           population_size=population,
           cai_on = True, 
           bai_on = True, 
           cpg_on = True,
           threads=20)
    pd.DataFrame(timefit).to_csv(filename,index=False)
    return(timefit)
    
```

```python
timefits = []
for comb in combinations:
    prefix=comb[0]
    aa_seq=comb[1]
    tissue=comb[2]
    population=comb[3]
    
    print(prefix+str(population),'\n')
    filename = prefix+str(population)+'.csv'
    timefit = run_pygad_variable_pop_mp.run_GA(aa_seq, 
           tissue, 
           population_size=population,
           cai_on = True, 
           bai_on = True, 
           cpg_on = True,
           threads=20)
    pd.DataFrame(timefit).to_csv(filename,index=False)
```
