```python
def create_table(df, species):
    organelle = 'genomic'
    df = df[(df['SPECIES']==species) & (df['ORGANELLE']==organelle)]
    rm_cols = ['TAXID','DIVISION','ASSEMBLY','SPECIES','ORGANELLE','TRANSLATION TABLE','# CDS','# CODON PAIRS','UNNAMED: 4104']
    df = df.drop(rm_cols,axis=1)
    file_nm = f'../../references/CoCoPUTs_codon_usage/bicodon_usage/{organelle}-{species}.bicodon.txt'.replace(' ','_')
    print(file_nm)
    df.to_csv(file_nm,index=False,sep='\t')
```

```python
df_rs = pd.read_csv(refseq,sep='\t')
df_rs.columns = [x.upper() for x in df_rs.columns]
df_rs
```

## Refseq vs CDS

```python
# rs_human = '/grid/home/nbourgeois/codonOpt/tissue_cocoputs/species/tables/Homo_sapiens.genomic.refseq.tsv'
# rs_human_weights = get_bicodon_weights(rs_human,default_loc=False)

rs_human='genomic-Homo_sapiens'
rs_human_weights = get_bicodon_weights(rs_human,default_loc=True)
default_human_weights = get_bicodon_weights('Genomic',default_loc=True)
```

```python
total_diff = 0
max_diff = 0
for key in default_human_weights.keys():
    diff = abs(default_human_weights[key] - rs_human_weights[key])
    total_diff += diff
    max_diff = max(max_diff,diff)

print(total_diff/len(default_human_weights.keys()), max_diff)
```
