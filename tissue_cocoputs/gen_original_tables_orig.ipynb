```python
cds = pd.read_csv('o586358-Human_CDS.tsv',sep='\t')
cols = cds.columns
cds = cds.drop(columns=['GGG'])
cds = cds.reset_index()
cds.columns = cols
cds = cds.drop(columns = ['Gene ID', 'Accession', 'Division', 'Assembly', 'Taxid',
                          'Species', 'Organelle', '# Codons', 'GC%', 'GC1%', 'GC2%', 'GC3%'])
cds = cds.groupby('Protein ID').sum()
cds
```

```python
mg = mygene.MyGeneInfo()
geneSyms = mg.querymany(cds.index.values, scopes='refseq', fields='ensembl.transcript', species='human')

```

```python
def get_weights(geneSyms):
        
    refseqs = []
    weights = []
    counter = 0
    for gdict in geneSyms:
        if 'ensembl' in gdict:
            for ensmbl in gdict['ensembl']:
                weight = liver_weight[geneSyms[3]['ensembl']['transcript']].sum()
                refseqs.append(gdict['query'])
                weights.append(weight)
        counter +=1
    return(refseqs,ensmbs)

refseqs,weights = get_weights(geneSyms)
```

```python
# liver_weight_sorted = liver_weight[ensmbs]
cds_sorted = cds.loc[refseqs,:]
cds_weighted = (cds_sorted.T * weights).sum(axis=1)
cds_weighted / sum(cds_weighted) * 1000
```
