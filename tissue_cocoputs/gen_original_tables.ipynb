```python
def convert(geneSyms):
        
    refseqs = []
    ensmbs = []
    counter = 0
    for gdict in geneSyms:
        if 'ensembl' in gdict:
            if len(gdict['ensembl'])==1:
                ensmbl = gdict['ensembl']['gene']
                if ensmbl in geneList:
                    refseqs.append(gdict['query'])
                    ensmbs.append(ensmbl)
            else:
                for ensmblg in gdict['ensembl']:
                    ensmbl = ensmblg['gene']
                    if ensmbl in geneList:
                        refseqs.append(gdict['query'])
                        ensmbs.append(ensmbl)
        counter +=1
    return(refseqs,ensmbs)

refseqs,ensmbs = convert(geneSyms)
```

```python
liver_weight_sorted = liver_weight[ensmbs]
cds_sorted = cds.loc[refseqs,:]
cds_weighted = (cds_sorted.T * liver_weight_sorted.values).sum(axis=1)
cds_weighted / sum(cds_weighted) * 1000
```
