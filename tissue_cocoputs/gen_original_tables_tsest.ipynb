```python

gtex_data = pd.read_csv('GTEx_head',sep='\t',skiprows=2)
gtex_cols = gtex_data.columns

gtex_data
```

```python
gtex_data['gene_id'] = gtex_data['gene_id'].str.split('.',expand=True)[0]
gtex_sum = gtex_data.groupby('gene_id').sum()
gtex_sum
```

```python

gtex_samp_data= pd.read_csv('GTEx_Analysis_v8_Annotations_SampleAttributesDS.txt',sep='\t')
liver_samp = gtex_samp_data.loc[gtex_samp_data['SMTS'] == 'Liver']['SAMPID'].values.tolist()
liver_cols = gtex_sum.columns[gtex_sum.columns.isin(liver_samp)]
gtex_sum_liver = gtex_sum[liver_cols]
liver_weight = gtex_sum_liver.mean(axis=1)
```

```python
mg = mygene.MyGeneInfo()
geneList = liver_weight.index.tolist()
geneSyms = mg.querymany(geneList , scopes='ensembl.gene', fields='symbol', species='human')
geneSym_list = [dict['symbol'] for dict in geneSyms]
```

```python
liver_weight_sym = liver_weight.copy()
liver_weight_sym.index = geneSym_list
liver_weight_sym
```

```python
cds = pd.read_csv('o586358-Human_CDS.tsv',sep='\t')
cols = cds.columns
cds = cds.drop(columns=['GGG'])
cds = cds.reset_index()
cds.columns = cols
cds = cds.set_index('Gene ID')
cds = cds.drop(columns = ['Protein ID', 'Accession', 'Division', 'Assembly', 'Taxid',
                          'Species', 'Organelle', '# Codons', 'GC%', 'GC1%', 'GC2%', 'GC3%'])
cds
```

```python
mut_ind = liver_weight_sym.index.intersection(cds.index)
liver_weight_sym = liver_weight_sym.loc[mut_ind]
cds = cds.loc[mut_ind,:]

cds = cds.drop(columns = ['Gene ID', 'Protein ID', 'Accession', 'Division', 'Assembly', 'Taxid',
                          'Species', 'Organelle', '# Codons', 'GC%', 'GC1%', 'GC2%', 'GC3%'])
```
